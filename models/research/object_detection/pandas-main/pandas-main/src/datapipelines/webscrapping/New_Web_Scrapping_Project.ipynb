```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(resp_data.text, 'html.parser') #Creating Beautiful Soup Object
```

```python
type(doc)
```

```python
animal_span_tags = doc.find_all('figure') # Finding Mammals in the home page
```

```python
len(animal_span_tags) # Found 15 animals
```

```python
animal_span_tags[6] # content of a Single Animal Html Page
```

```python
animal_href_tags = doc.find_all('a', href = lambda L: L and L.startswith('/species')) # Extracting Hyperlink for eeach animal individually
```

```python
print(animal_href_tags[0])
```

```python
print(animal_href_tags[0]['href']) # Extracting Animal URL
```

```python
animal_url_tags = []  # creating individual animal complete URLs
for i in range(40):
    animal_url_tags.append("https://www.worldwildlife.org" + animal_href_tags[i]['href'])
```

```python
print(animal_url_tags[:-1]) #checking contents of the URL list

len(animal_url_tags)
```
